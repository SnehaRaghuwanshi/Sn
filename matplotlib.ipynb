```python
# graphs with more values on dots with different types of line and color
x_axis=np.array([1,2,3,4,5])
y_axis=np.array([10,13,16,18,20])
plt.plot(x_axis,y_axis,'o:r')# o means marker,: means line will be dotted,r means red color
plt.show
```

```python
# graph with different marker size and marker color etc
x_axis=np.array([1,2,3,4,5])
y_axis=np.array([10,13,16,18,20])
plt.plot(x_axis,y_axis,marker = 'o',ms= 10,mfc = 'r',mec = 'g') #ms means marker size,mfc means inner color,mec means outer color
plt.show
```

```python
# graph with with colors from color palet 
#for color palet go to w3 schools and search color picker and then copy color code from Hex
x_axis=np.array([1,2,3,4,5])
y_axis=np.array([10,13,16,18,20])
plt.plot(x_axis,y_axis,marker = '*',ms = 20 ,mec = '#d926d9', mfc = '#ffb3ff' )
plt.show
```

```python
# when you work only on 1 axis 
y_axis = np.array([12,15,19,20,11])
plt.plot(y_axis,linestyle = ':',) #mentioning linestyle is important otherwise it will take it as y axis  
plt.show
```

```python
y_axis = np.array([12,15,19,20,11])
plt.plot(y_axis,ls = ':') #ls is short form of linestyle
```

```python
y_axis=np.array([12,15,20,17,18])
plt.plot(y_axis,color='orange')
```

```python
y_axis=np.array([12,13,14,18,1,0])
plt.plot(y_axis,linewidth='20')# use linewidth or the shorter lw to change the width of the line.
plt.show
```

***SUBPLOTTING*** 

```python
#for subplot we use keyword subplot()
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 6)#this means 2 rows, 3 columns, and this plot is the sixth plot.
plt.plot(x,y)
#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)#this means 2 rows, 3 columns, and this plot is the second plot.
plt.plot(x,y)
#plot 3
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)#this means 2 rows, 3 columns, and this plot is the third plot.
plt.plot(x,y)
#plot 4
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)##this means 2 rows, 3 columns, and this plot is the fourth plot.
plt.plot(x,y)
#plot 5
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)#this means 2 rows, 3 columns, and this plot is the fifthplot.
plt.plot(x,y)
#plot 6
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 1)#this means 2 rows, 3 columns, and this plot is the first plot.
plt.plot(x,y)

plt.show()
```
